```python
result = []
for i in range(df.shape[0]): 
    if '-' in df['4'][i]:
            a, b = df['4'][i].split('-')
            a, b = int(a), int(b)
            result.append(list(range(a, b + 1)))
    else:
            a = int(df['4'][i])
            result.append(a)
```
